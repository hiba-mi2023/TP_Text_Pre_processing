```python
import pandas as pd
import re
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer

# Download the 'stopwords' dataset
import nltk
nltk.download('stopwords')

# Remove stopwords
stop_words = set(stopwords.words('english'))
df['text_cleaned'] = df['text_cleaned'].apply(lambda x: ' '.join([word for word in x.split() if word not in stop_words]))
```

```python
freq_words = pd.Series(' '.join(df['text_cleaned']).split()).value_counts()[:10]
df['text_cleaned'] = df['text_cleaned'].apply(lambda x: ' '.join([word for word in x.split() if word not in freq_words]))
```

```python
rare_words = pd.Series(' '.join(df['text_cleaned']).split()).value_counts()[-10:]
df['text_cleaned'] = df['text_cleaned'].apply(lambda x: ' '.join([word for word in x.split() if word not in rare_words]))
```

```python
stemmer = PorterStemmer()
df['text_cleaned'] = df['text_cleaned'].apply(lambda x: ' '.join([stemmer.stem(word) for word in x.split()]))
```

```python
import pandas as pd
import re
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer

# Download the 'wordnet' dataset
import nltk
nltk.download('wordnet')


# Lemmatization
lemmatizer = WordNetLemmatizer()
df['text_cleaned'] = df['text_cleaned'].apply(lambda x: ' '.join([lemmatizer.lemmatize(word) for word in x.split()]))

```

```python
df['text_cleaned'] = df['text_cleaned'].apply(lambda x: re.sub(r'(?i)\b((?:https?://|www\d?\d?\d?\d?\d?\d?\d?\d?\.\w+\.?[a-z]{2,4}/)(?:[^\s()<>{}\[\]]+|\([^\s()]*?\([^\s()]+\)[^\s()]*?\)|\([^\s]+?\))+(?:\([^\s()]*?\([^\s()]+\)[^\s()]*?\)|\([^\s]+?\)|[^\s`!()\[\]{};:\'".,<>?«»“”‘’]))', '', x))
df['text_cleaned'] = df['text_cleaned'].apply(lambda x: re.sub(r'<.*?>', '', x))
```
